```python
import os
# !pip install -q ucimlrepo
from ucimlrepo import fetch_ucirepo
```

```python
# fetch dataset
mushroom = fetch_ucirepo(id=73)

# data (as pandas dataframes)
X = mushroom.data.features
y = mushroom.data.targets

# metadata
#mushroom.metadata
```

```python
import pandas as pd
import matplotlib.pyplot as plt

original_df = pd.concat([X, y], axis=1)
# original_df = original_df.drop(["stalk-root"], axis=1)
```

```python
original_df
```

```python
# two separate dataframes for comparative figures
edible_df = original_df[original_df['poisonous'] == "e"]
poison_df = original_df[original_df['poisonous'] == "p"]
```

```python
# counts of poisonous v edible
PvE = original_df['poisonous'].value_counts().plot(kind = "bar",
                                                  title = "Poisonous v. Edible",
                                                  ylabel = "# of species",
                                                  color = ["darkblue", "red"])
plt.tight_layout()
plt.savefig("static/assets/visuals/poison_vs_eds.png")
plt.show()
```
